```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from example import Model

X = np.linspace(-3, 5)
y = X ** 2 + 1

X = X.reshape(len(X), 1)
y = y.reshape(len(y), 1)

print(X[0])
print(y[0])
```

```python
plt.scatter(X, y ,color='red')
plt.show()
```

```python
m = Model(1, 1)
m.train(X, y, epochs=5000, learning_rate=1e-3)
```

```python
trained = [m.forward(x)[0].val for x in X]

plt.scatter(X, y ,color='red')
plt.plot(X, trained, color='blue')
plt.show()
```
