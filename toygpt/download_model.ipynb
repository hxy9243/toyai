```python
from transformers import GPT2LMHeadModel

model_hf = GPT2LMHeadModel.from_pretrained('gpt2')
sd_hf = model_hf.state_dict()

for k, v in sd_hf.items():
    print(k, v.shape)
```

```python
from transformers import GPT2Tokenizer

encoding = GPT2Tokenizer.from_pretrained('gpt2')
encoded = encoding.encode('Hello world! We are learning the', return_tensors='pt')

output = model_hf.generate(encoded, max_length=50, num_return_sequences=1)
print(
    encoding.decode(output[0], skip_special_tokens=True)
)
```

```python
with torch.no_grad():
    forwarded = model_hf.forward(encoded)

print(encoded.shape)
print(forwarded[0].shape)
```
